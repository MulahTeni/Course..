# Data Visualization
## Using Seaborn and Matplotlib
There are a few useful visualization techniques below. It's just for training. Data is massive!

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
# for time related data
import matplotlib.pyplot as plt
%matplotlib inline
# for cell plot
import seaborn as sns
```

Ad datapath to os

```python
import os
# if not os.path.exists("../cancer.csv"):
#     os.symlink("../cancer-data/cancer.csv", "../input/cancer.csv")
```

Read and load data
Drop categorical columns for plotting

```python
data_path = "cancer.csv"
f_data = pd.read_csv(data_path, index_col = 'Id')
_data = f_data[['Diagnosis', 'Radius (mean)', 'Area (mean)']]
```

# Visualization variations
## - Line Plot

```python
plt.figure(figsize=(6, 6))
sns.lineplot(data = _data)
plt.title("Line Plot")
```

Plot a part of data

```python
sns.lineplot(data=_data['Area (mean)'])
plt.title("Part Line Plot")
```

## - Bar Plot

```python
plt.figure(figsize=(30,8))
sns.barplot(x=_data.index, y=_data['Area (mean)'])
plt.title("Bar Plot")
plt.ylabel("Area")
```

## - Heatmap Plot

```python
plt.figure(figsize=(20, 6))
plt.title("Heatmap Plot")
sns.heatmap(data=_data[['Area (mean)', 'Radius (mean)']] , annot = True)
plt.xlabel("ID")
```

## - Scatter Plot

```python
sns.scatterplot(x=_data['Radius (mean)'], y=_data['Area (mean)'])
```

## - Scatter Plot with Regression Line

```python
sns.regplot(x=_data['Radius (mean)'], y=_data['Area (mean)'], color="red", marker="*")
```

## - Scatter Plot with color-coding

```python
sns.scatterplot(x=_data['Radius (mean)'], y=_data['Area (mean)'], hue=_data['Diagnosis'])
```

## - Scatter Plot with color-coding and 2 Regression Lines

```python
sns.lmplot(x="Radius (mean)", y="Area (mean)", hue="Diagnosis", data=_data)
```

## - Scatter Plot, split

```python
sns.swarmplot(x=_data['Radius (mean)'], y=_data['Area (mean)'])
# Massive data error
```

```python
sns.stripplot(x=_data['Radius (mean)'], y=_data['Area (mean)'])
# Run without data error
```

## - Histogram Plot

```python
sns.histplot(data=_data, x='Area (mean)', hue='Diagnosis')
```

## - Joint Plot

```python
sns.jointplot(data=_data, x="Radius (mean)", y="Area (mean)")
```

## - Joint Plot with color-coding

```python
sns.jointplot(data=_data, x="Radius (mean)", y="Area (mean)", hue="Diagnosis")
```

## - KDE Plot

```python
sns.kdeplot(data=_data, x="Area (mean)")
```
